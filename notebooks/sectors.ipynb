```python
from datetime import datetime
from pathlib import Path

import pandas as pd

from cuery import task
from cuery.work import DirceJobs, DirceTasks

GDRIVE = Path("~/Google Drive/Shared drives/Solutions").expanduser()
PROJECT = GDRIVE / "Research/future_of_work/"
```

```python
DATA_FP = PROJECT / "inputs/ine_dirce_aggregated_by_activity.parquet"

data = pd.read_parquet(DATA_FP)
data.columns = [c.replace(" ", "_") for c in data.columns]
data = data.rename(columns={"Division": "sector", "Actividad_principal": "subsector"})
data = data.sort_values("Estimated_Employees_2024", ascending=False)
display(data)

n_examples = 5
df = data.iloc[:5]
```

```python
chain = task.Chain(DirceJobs, DirceTasks)
result = await chain(df, model="openai/gpt-4.1-mini", n_concurrent=20)
result
```

```python
chain._usage[["prompt", "completion", "cost"]].sum()
```

```python
now = datetime.now().isoformat()
out_fnm = PROJECT / f"outputs/ine_dirce_ai_tasks_{now}.csv"
result.to_csv(out_fnm, index=False)
```
